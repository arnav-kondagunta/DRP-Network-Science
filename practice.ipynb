```python
import networkx as nx
import matplotlib.pyplot as plt
```

Creating and Editing Graphs

```python
G = nx.Graph()
```

```python
#adding nodes
G.add_node(1) #one at a time
G.add_nodes_from([2, 3]) #as a list
G.add_nodes_from([(4, {"color": "red"}), (5, {"color": "green"})]) #with attributes

H = nx.path_graph(10) 
G.add_nodes_from(H) #G contains nodes from H
G.add_node(H) #adds H as a node
```

```python
#adding edges
G.add_edge(1, 2) #one at a time
e = (1, 2)
G.add_edge(*e) #via tuple
G.add_edges_from([(1, 2), (1, 3)]) #list of edges
G.add_edges_from(H.edges)
```

```python
#ignores edges that are already present
G.clear()
G.add_edges_from([(1, 2), (1, 3)])
G.add_node(1)
G.add_edge(1, 2)
G.add_node("spam")        # adds node "spam"
G.add_nodes_from("spam")  # adds 4 nodes: 's', 'p', 'a', 'm'
G.add_edge(3, 'm')
```

```python
G.number_of_edges()
```

```python
G.number_of_nodes()
```

```python
#directed graph
DG = nx.DiGraph()
DG.add_edge(2, 1)
DG.add_edge(1, 3)
DG.add_edge(2, 4)
DG.add_edge(1, 2)
assert list(DG.successors(2)) == [1, 4]
assert list(DG.edges) == [(2, 1), (2, 4), (1, 3), (1, 2)]
```

Drawing Graphs from the Textbook

```python
#Figure 2.13, pg. 45
G.clear()
G.add_nodes_from(["A", "B", "C", "D", "E", "F"])
e = [("A", "B"), ("B", "C"), ("B", "D"), ("F", "C"), ("F", "E"), ("A", "E"), ("C", "D")] 
G.add_edges_from(e)
subax1 = plt.subplot()
pos = {
    "A" : (0.7, 1),
    "B" : (0.7, 0.5),
    "C" : (0.35, 0),
    "D" : (1, 0),
    "E" : (0, 1),
    "F" : (0, 0.5)
 
}
nx.draw(G, pos, with_labels=True, font_weight='bold')
```
